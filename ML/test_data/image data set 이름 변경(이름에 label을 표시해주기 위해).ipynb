```python
import os
path="Test_Data"
os.listdir(path)
file_list=os.listdir(path)[:10]
file_list
```

```python
for i in file_list:
    temp_path=os.path.join(path,i)
    print(temp_path)
   # print(os.listdir(temp_path))
    real_path=os.listdir(temp_path)
    for j in real_path:
        print(temp_path+"/"+j)
        os.rename(temp_path+"/"+j,temp_path+"/"+str(i)+"_"+j)
        
```
